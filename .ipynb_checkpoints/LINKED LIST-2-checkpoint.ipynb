### REVERSE LINKED LIST-1

```python
class node:
    def __init__(self,data):
        self.data=data
        self.next=None
        
def take_input():
    input_list=[int(x) for x in input().split()]
    head=None
    tail=None
    for curr_data in input_list:
        if curr_data==-1:
            break
            
        new_node=node(curr_data)
        if head is None:
            head=new_node
            tail=new_node     
        else:
            tail.next=new_node
            tail=new_node
            
    return head,tail

def printLL(head):
    while head is not None:
        print(str(head.data),end=' ')
        head=head.next
    return

def reverse_ll1(head):
    if head.next==None or head is None:
        return head
    
    smallhead=reverse_ll1(head.next)
    curr=smallhead
    while curr.next is not None:
        curr=curr.next
    tail=curr
    tail.next=head
    head.next=None
    return smallhead

head,tail=take_input()
headc=reverse_ll1(head)
printLL(headc)
```

### REVERSE LINKED LIST-ITERATIVELY

```python
class node:
    def __init__(self,data):
        self.data=data
        self.next=None
        
def take_input():
    input_list=[int(x) for x in input().split()]
    head=None
    tail=None
    for curr_data in input_list:
        if curr_data==-1:
            break
            
        new_node=node(curr_data)
        if head is None:
            head=new_node
            tail=new_node     
        else:
            tail.next=new_node
            tail=new_node
            
    return head,tail

def printLL(head):
    while head is not None:
        print(head.data,end=' ')
        head=head.next    

def  lenll(head):
    count=0
    while head is not None:
        count=count+1
        head=head.next
    return count

def reverse_lli(head):
    i,count,lll=1,1,lenll(head)
    rcurr=tail
    while i<lll:
        curr=head
        for j in range(0,lll-count-1,1):
            curr=curr.next
        rcurr.next=curr
        rcurr=rcurr.next
        count+=1
        i+=1
    rcurr.next=None
    return tail

head,tail=take_input()
print(tail.next)
headc=reverse_lli(head)
printLL(headc)
```

### MID POINT OF LINKED LIST

```python
class node:
    def __init__(self,data):
        self.data=data
        self.next=None
        
def take_input():
    input_list=[int(x) for x in input().split()]
    head=None
    tail=None
    for curr_data in input_list:
        if curr_data==-1:
            break
            
        new_node=node(curr_data)
        if head is None:
            head=new_node
            tail=new_node     
        else:
            tail.next=new_node
            tail=new_node
            
    return head,tail

def  lenll(head):
    count=0
    while head is not None:
        count=count+1
        head=head.next
    return count

def mid_ll(head):
    lll=lenll(head)
    count=0
    if lll%2==0:
        c=lll//2-1
    else:
        c=lll//2
    while count<c:
        head=head.next
        count+=1
    print(head.data)
    return

head,tail=take_input()
mid_ll(head)
```

### MERGE TWO SORTED LINKED LIST

```python
class node:
    def __init__(self,data):
        self.data=data
        self.next=None
        
def take_input():
    input_list=[int(x) for x in input().split()]
    head=None
    tail=None
    for curr_data in input_list:
        if curr_data==-1:
            break
            
        new_node=node(curr_data)
        if head is None:
            head=new_node
            tail=new_node     
        else:
            tail.next=new_node
            tail=new_node
            
    return head,tail

def printLL(head):
    while head is not None:
        print(str(head.data),end=' ')
        head=head.next
    return

def merge_ll(head1,head2):
    if head1.data<head2.data:
        new_head=head1
        curr=new_head
        head1=head1.next
    else:
        new_head=head2
        curr=new_head
        head2=head2.next
    
    while head1 is not None and head2 is not None:
            if head1.data<head2.data:
                temp=head1
                curr.next=temp
                head1=head1.next            
            else:
                temp=head2
                curr.next=temp
                head2=head2.next
            curr=curr.next
    while head1 is None and head2 is not None:
        temp=head2
        curr.next=temp
        head2=head2.next
        curr=curr.next
    
    while head1 is not None and head2 is None:
        temp=head1
        curr.next=temp
        head1=head1.next
        curr=curr.next
        
    return new_head               

head1,tail1=take_input()
head2,tail2=take_input()
headc=merge_ll(head1,head2)
printLL(headc)
```

### MERGE SORT

```python
class node:
    def __init__(self,data):
        self.data=data
        self.next=None
        
def take_input():
    input_list=[int(x) for x in input().split()]
    head=None
    tail=None
    for curr_data in input_list:
        if curr_data==-1:
            break
            
        new_node=node(curr_data)
        if head is None:
            head=new_node
            tail=new_node     
        else:
            tail.next=new_node
            tail=new_node
            
    return head,tail

def printLL(head):
    while head is not None:
        print(str(head.data),end=' ')
        head=head.next
    return

def  lenll(head):
    count=0
    while head is not None:
        count=count+1
        head=head.next
    return count

def merge_ll(head1,head2):
    if head1.data<head2.data:
        new_head=head1
        curr=new_head
        head1=head1.next
    else:
        new_head=head2
        curr=new_head
        head2=head2.next
    
    while head1 is not None and head2 is not None:
            if head1.data<head2.data:
                temp=head1
                curr.next=temp
                head1=head1.next            
            else:
                temp=head2
                curr.next=temp
                head2=head2.next
            curr=curr.next
    while head1 is None and head2 is not None:
        temp=head2
        curr.next=temp
        head2=head2.next
        curr=curr.next
    
    while head1 is not None and head2 is None:
        temp=head1
        curr.next=temp
        head1=head1.next
        curr=curr.next
        
    return new_head

def merge_sortll(head,tail,lll):
    if head is None:
        return head
        
    c=lll//2
    count,curr=0,head
    
    while count<c:
        curr=curr.next
        count+=1
    
    h2=curr
    h2=h2.next
    curr.next=None
    
    if c%2==0:
        merge_sortll(h2,tail,c-1)
        merge_sortll(head,curr,c+1)        
    else:
        merge_sortll(h2,tail,c)
        merge_sortll(head,curr,c+1)        
    
    merge_ll(head,h2)
    
    curr.next=h2
    
    return head
     
head,tail=take_input()
lll=lenll(head)
headc=merge_sortll(head,tail,lll)
printLL(headc)
```

### FIND A NODE IN A LINKED LIST-RECURSIVELY

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

def linearSearchRecursive(head, n,i):
    if head==None:
        return -1
    
    if head.data==n:
        return i
    
    position=linearSearchRecursive(head.next, n,i+1)
    
    return position
    
def ll(arr):
    if len(arr)==0:
        return None
    head = Node(arr[0])
    last = head
    for data in arr[1:]:
        last.next = Node(data)
        last = last.next
    return head

# Main
from sys import setrecursionlimit
setrecursionlimit(11000)
# Read the link list elements including -1
arr=list(int(i) for i in input().strip().split(' '))
# Create a Linked list after removing -1 from list
l,i = ll(arr[:-1]),0
data=int(input())
index = linearSearchRecursive(l, data,i)
print(index)
```
