### FIND A NODE IN A LINKED LIST-RECURSIVELY

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

def linearSearchRecursive(head, n,i):
    if head==None:
        return -1
    
    if head.data==n:
        return i
    
    position=linearSearchRecursive(head.next, n,i+1)
    
    return position
    
def ll(arr):
    if len(arr)==0:
        return None
    head = Node(arr[0])
    last = head
    for data in arr[1:]:
        last.next = Node(data)
        last = last.next
    return head

# Main
from sys import setrecursionlimit
setrecursionlimit(11000)
# Read the link list elements including -1
arr=list(int(i) for i in input().strip().split(' '))
# Create a Linked list after removing -1 from list
l,i = ll(arr[:-1]),0
data=int(input())
index = linearSearchRecursive(l, data,i)
print(index)
```

### K REVERSE

```python
class node:
    def __init__(self,data):
        self.data=data
        self.next=None
        
def take_input():
    input_list=[int(x) for x in input().split()]
    head=None
    tail=None
    for curr_data in input_list:
        if curr_data==-1:
            break
            
        new_node=node(curr_data)
        if head is None:
            head=new_node
            tail=new_node     
        else:
            tail.next=new_node
            tail=new_node
            
    return head,tail


def print_reverse(head,tail2):
    if head==tail2:
        print(head.data,end=' ')
        return
    print_reverse(head.next,tail2)
    print(head.data,end=' ')
    return

def  lenll(head):
    count=0
    while head is not None:
        count=count+1
        head=head.next    
    return count

def k_reverse(head,tail,k):
    lll=lenll(head)
    if k>lll:
        print_reverse(head,tail)
    else:
        thead,ttail=head,head
        while thead is not None:
            count=1
            while count<k:
                if ttail.next==None:
                    break
                else:
                    ttail=ttail.next
                    count+=1           
            print_reverse(thead,ttail)
            ttail=ttail.next
            thead=ttail           
    return    

head,tail=take_input()
k=int(input())
headc=k_reverse(head,tail,k)
```
