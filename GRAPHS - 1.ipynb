## ALL CONNECTED COMPONENTS - USING DFS

```python
# 1. DFS
# 2. BFS
import queue
class Graph:
    def __init__(self,nVertices):
        self.vertices=nVertices
        self.adjacentMatrix=[[0 for j in range(nVertices)] for i in range(nVertices)]
        
    def addEdge(self,v1,v2):
        self.adjacentMatrix[v1][v2]=1
        self.adjacentMatrix[v2][v1]=1
        return
        
    def removeEdge(self,v1,v2):
        if not(self.containsEdge):
            return 
        self.adjacentMatrix[v1][v2]=0
        self.adjacentMatrix[v2][v1]=0
        return
    
    def __connectedComponentsHelper(self,i,visited,li):
        li.append(i)
        visited[i]=True
        for k in range(self.vertices):
            if self.adjacentMatrix[i][k]==1 and visited[k] is False:
                self.__connectedComponentsHelper(k,visited,li)
        return         
    
    def connectedComponents(self):
        visited=[False]*(self.vertices)
        li_final=[]        
        for i in range(self.vertices):
            if visited[i] is False:
                li=[]
                self.__connectedComponentsHelper(i,visited,li)
                li_final.append(li)        
        return li_final
         
    def containsEdge(self,v1,v2):
        return self.adjacentMatrix[v1][v2]==1
    
    def __str__(self):
        return str(self.adjacentMatrix)

V,E=[int(x) for x in input().split()]
g=Graph(V)
for i in range(E):
    v1,v2=[int(x) for x in input().split()]
    g.addEdge(v1,v2)

li=g.connectedComponents()
for row in li:
    row.sort()
    print(*row)
    
```
