# IMPROVED POWER CODE

```python
def power(x, n):
    if n==0:
        return 1
    
    smalloutput=power(x,n//2)
    
    if n%2==0:
        return smalloutput*smalloutput
    elif n%2!=0:
        return x*smalloutput*smalloutput
    

# Main
from sys import setrecursionlimit
setrecursionlimit(11000)
x, n=list(int(i) for i in input().strip().split(' '))
print(power(x, n))
```

### IMPROVED EQUILIBRIUM INDEX

```python
def equilibriumIndex(arr,sum):
    lsum=0
    rsum=sum
    for i in range (len(arr)):
        lsum=sum-rsum
        rsum=rsum-arr[i]
        if (lsum==rsum):
            return i
    else:
        return -1

# Main
n = int(input())
arr = [int(i) for i in input().strip().split()]
sum=0
for i in range(len(arr)):
    sum=sum+arr[i]
print(equilibriumIndex(arr,sum))
```

```python
def merge(l1,l2,l):
    k=0
    i,j=0,0    
    while(i<len(l1) and j<len(l2)):
        if l1[i]<=l2[j]:
            l[k]=l1[i]
            i=i+1
            k=k+1
        elif l2[j]<l1[i]:
            l[k]=l2[j]
            j=j+1
            k=k+1
    o=i
    p=j
    if(o<=len(l1)-1):
        for c in range(o,len(l1)):
            l[k]=l1[c]
            k=k+1
    elif(p<=len(l2)-1):
        for d in range(p,len(l2)):
            l[k]=l2[d]
            k=k+1    

def merge_sort(li):
    g=len(li)
    if  g==0 or g==1:
        return li
    
    mid=len(li)//2
        
    s1=li[0:mid]
    s2=li[mid:]
    
    merge_sort(s1)
    merge_sort(s2)
    
    merge(s1,s2,li)
    
def trip_sum(arr,y):
    for st in range(len(arr)):
        i=st+1
        j=len(arr)-1
        x=y-arr[st]       
        while(i<j):
            if arr[i]+arr[j]<x:
                i=i+1
            elif arr[i]+arr[j]>x:
                j=j-1
            else:
                if arr[i]==arr[j]:
                    k=i
                    m=1
                    while k<(len(arr)-1):
                        if arr[k]==arr[k+1]:
                            m=m+1
                            k=k+1
                        else:
                            break
                    q=(m*(m-1))//2
                    for r in range(1,q+1,1):
                        print(arr[st],arr[i],arr[j])
                    i=i+m
                    j=j-m
                
                else:
                    k=i
                    m=1
                    while k<(len(arr)-1):
                        if arr[k]==arr[k+1]:
                            m=m+1
                            k=k+1
                        else:
                            break
                    l=j
                    u=1
                    while l>=0:
                        if arr[l]==arr[l-1]:
                            u=u+1
                            l=l-1
                        else:
                            break
                    q=m*u
                    for r in range(1,q+1,1):
                        print(arr[st],arr[i],arr[j])
                    i=i+m
                    j=j-u    
    
n=int(input())
arr=[int(x) for x in input().split()]
y=int(input())
merge_sort(arr)
trip_sum(arr,y)
```
