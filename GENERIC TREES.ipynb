### PRINT TREE

```python
class TreeNode:
    def __init__(self,data):
        self.data=data
        self.children=list()
        
def printTree(root):
    #THIS IS AN EDGE CASE NOT A BASE CASE
    if root==None:
        return
    print(root.data,end=' ')
    for child in root.children:
        printTree(child)
    return
        
n1=TreeNode(5)
n2=TreeNode(2)
n3=TreeNode(9)
n4=TreeNode(8)
n5=TreeNode(7)
n6=TreeNode(15)
n7=TreeNode(1)

n1.children.append(n2)
n1.children.append(n3)
n1.children.append(n4)
n1.children.append(n5)
n3.children.append(n6)
n3.children.append(n7)

printTree(n1)
```

### NUMBER OF NODES

```python
class TreeNode:
    def __init__(self,data):
        self.data=data
        self.children=list()
        
def printTreeDetailed(root):
    #THIS IS AN EDGE CASE NOT A BASE CASE
    if root==None:
        return
    
    print(str(root.data)+':',end='')
    for child in root.children:
        print(child.data,end=',')
        
    print()
    
    for child in root.children:
        printTreeDetailed(child)
        
def TreeInput():
    print("ENTER ROOT DATA")
    root_data=int(input())
    if root_data==-1:
        return None
    
    root=TreeNode(root_data)
    
    print("ENTER THE NUMBER OF CHILDREN ROOT HAS")
    childrenCount=int(input())
    for i in range(childrenCount):
        child=TreeInput()
        root.children.append(child)
    return root  

def numNodes(root):
    # THIS IS AN EDGE CASE NOT BASE CASE
    if root ==None:
        return 0
    
    count=1
    for child in root.children:
        count=count+numNodes(child)
    return count    
        
root=TreeInput()
printTreeDetailed(root)
print(numNodes(root))
```

### SUM OF ALL NODES

```python
class treeNode:
    def __init__(self, data):
        self.data = data
        self.children = []
    def __str__(self):
        return str(self.data)

def sumofNodes(tree):
    # THIS IS EDGE CASE NOT BASE CASE
    if tree==None:
        return 0
    
    sum=tree.data
    for child in tree.children:
        sum=sum+sumofNodes(child)
    return sum    

def createLevelWiseTree(arr):
    root = treeNode(int(arr[0]))
    q = [root]
    size = len(arr)
    i = 1
    while i<size:
        parent = q.pop(0)
        childCount = int(arr[i])
        i += 1
        for j in range(0,childCount):
            temp = treeNode(int(arr[i+j]))
            parent.children.append(temp)
            q.append(temp)
        i += childCount
    return root

# Main
arr = list(int(x) for x in input().strip().split(' '))
tree = createLevelWiseTree(arr)
print(sumofNodes(tree))
```

### NODE WITH LARGEST DATA

```python
class treeNode:
    def __init__(self, data):
        self.data = data
        self.children = []
    def __str__(self):
        return str(self.data)

def maxDataNode(tree):
    # THIS IS EDGE CASE NOT BASE CASE
    if tree==None:
        return None    
    
    maxm=0
    for child in tree.children:
        child_max=maxDataNode(child)
        maxm=max(maxm,child_max)            
         
    maxm=max(maxm,tree.data)
    
    return maxm    

def createLevelWiseTree(arr):
    root = treeNode(int(arr[0]))
    q = [root]
    size = len(arr)
    i = 1
    while i<size:
        parent = q.pop(0)
        childCount = int(arr[i])
        i += 1
        for j in range(0,childCount):
            temp = treeNode(int(arr[i+j]))
            parent.children.append(temp)
            q.append(temp)
        i += childCount
    return root

# Main
arr = list(int(x) for x in input().strip().split(' '))
tree = createLevelWiseTree(arr)
print(maxDataNode(tree))
```

### TREE INPUT LEVELWISE

```python
class TreeNode:
    def __init__(self,data):
        self.data=data
        self.children=list()
        
def printTreeDetailed(root):
    #THIS IS AN EDGE CASE NOT A BASE CASE
    if root==None:
        return
    
    print(str(root.data)+':',end='')
    for child in root.children:
        print(child.data,end=',')
        
    print()
    
    for child in root.children:
        printTreeDetailed(child)
import queue        
def TreeInputLevelwise():
    q=queue.Queue()
    print('ENTER ROOT DATA')
    root_data=int(input())
    if root_data==-1:
        return None
    
    root=TreeNode(root_data)
    q.put(root)
    while(not(q.empty())):
        current_node=q.get()
        print("ENTER THE NUMBER OF CHILDREN OF ",current_node.data)
        current_children=int(input())
        for i in range(current_children):
            print('ENTER THE NEXT CHILD OF ',current_node.data)
            child_data=int(input())
            if child_data==-1:
                return None
            child=TreeNode(child_data)
            current_node.children.append(child)
            q.put(child)
    return root        
        
root=TreeInputLevelwise()
printTreeDetailed(root)        
```

### HEIGHT OF GENERIC TREE

```python
class treeNode:
    def __init__(self, data):
        self.data = data
        self.children = []
    def __str__(self):
        return str(self.data)

def height(tree):
    # THIS IS EDGE CASE NOT BASE CASE
    if tree==None:
        return 0
    
    maxm=0
    for child in tree.children:
        childh=height(child)
        maxm=max(maxm,childh)
            
    return 1+maxm
    
def createLevelWiseTree(arr):
    root = treeNode(int(arr[0]))
    q = [root]
    size = len(arr)
    i = 1
    while i<size:
        parent = q.pop(0)
        childCount = int(arr[i])
        i += 1
        for j in range(0,childCount):
            temp = treeNode(int(arr[i+j]))
            parent.children.append(temp)
            q.append(temp)
        i += childCount
    return root

# Main
arr = list(int(x) for x in input().strip().split(' '))
tree = createLevelWiseTree(arr)
print(height(tree))
```

### CONTAINS X

```python
class treeNode:
    def __init__(self, data):
        self.data = data
        self.children = []
    def __str__(self):
        return str(self.data)

def containsX(tree, x):
    if tree==None:
        return False
    
    if tree.data==x:
        return True
    
    for child in tree.children:
        stm=containsX(child,x)
        if stm:
            return True
    return False    
        

def createLevelWiseTree(arr):
    root = treeNode(int(arr[0]))
    q = [root]
    size = len(arr)
    i = 1
    while i<size:
        parent = q.pop(0)
        childCount = int(arr[i])
        i += 1
        for j in range(0,childCount):
            temp = treeNode(int(arr[i+j]))
            parent.children.append(temp)
            q.append(temp)
        i += childCount
    return root

# Main
x=int(input())
arr = list(int(x) for x in input().strip().split(' '))
tree = createLevelWiseTree(arr)
if containsX(tree,x):
    print('true')
else:
    print('false')
```

### COUNT LEAF NODES

```python
class treeNode:
    def __init__(self, data):
        self.data = data
        self.children = []
    def __str__(self):
        return str(self.data)

def leafNodeCount(tree):
    if tree==None:
        return 0
    
    if len(tree.children)==0:
        return 1
    
    count=0
    for child in tree.children:
        child_leaf=leafNodeCount(child)
        count=count+child_leaf
        
    return count    

def createLevelWiseTree(arr):
    root = treeNode(int(arr[0]))
    q = [root]
    size = len(arr)
    i = 1
    while i<size:
        parent = q.pop(0)
        childCount = int(arr[i])
        i += 1
        for j in range(0,childCount):
            temp = treeNode(int(arr[i+j]))
            parent.children.append(temp)
            q.append(temp)
        i += childCount
    return root

# Main
arr = list(int(x) for x in input().strip().split(' '))
tree = createLevelWiseTree(arr)
print(leafNodeCount(tree))
```

### STRUCTURALLY IDENTICAL

```python
class treeNode:
    def __init__(self, data):
        self.data = data
        self.children = []

def isIdentical(tree1, tree2):
    if tree1.data!=tree2.data:
        return False
    
    if (len(tree1.children)!=len(tree2.children)):
        return False
    else:
        for i in range(len(tree1.children)):
            stm=isIdentical(tree1.children[i],tree2.children[i])
            if stm==False:
                return False
            
    return True        
          
def createLevelWiseTree(arr):
    root = treeNode(int(arr[0]))
    q = [root]
    size = len(arr)
    i = 1
    while i<size:
        parent = q.pop(0)
        childCount = int(arr[i])
        i += 1
        for j in range(0,childCount):
            temp = treeNode(int(arr[i+j]))
            parent.children.append(temp)
            q.append(temp)
        i += childCount
    return root

# Main
arr1 = list(int(x) for x in input().strip().split(' '))
tree1 = createLevelWiseTree(arr1)
arr2 = list(int(x) for x in input().strip().split(' '))
tree2 = createLevelWiseTree(arr2)
if isIdentical(tree1, tree2):
    print('true')
else:
    print('false')
```

### REPLACE WITH DEPTH

```python
class treeNode:
    def __init__(self, data):
        self.data = data
        self.children = []

def replacewithDepth(tree,k):
    if tree==None:
        return 
    
    tree.data=k
    for child in tree.children:
        replacewithDepth(child,k+1)
        
    return    

def createLevelWiseTree(arr):
    root = treeNode(int(arr[0]))
    q = [root]
    size = len(arr)
    i = 1
    while i<size:
        parent = q.pop(0)
        childCount = int(arr[i])
        i += 1
        for j in range(0,childCount):
            temp = treeNode(int(arr[i+j]))
            parent.children.append(temp)
            q.append(temp)
        i += childCount
    return root

def printLevelAtNewLine(tree):
    q = [tree]
    newq = []
    while q:
        parent = q.pop(0)
        print(parent.data, end=' ')
        for child in parent.children:
            newq.append(child)
        if len(q)==0:
            q = newq
            newq = []
            print()  # Move to next Line

# Main
arr = list(int(x) for x in input().strip().split(' '))
tree = createLevelWiseTree(arr)
replacewithDepth(tree,0)
printLevelAtNewLine(tree)
```

### NODE WITH MAXIMUM CHILD SUM

```python
class treeNode:
    def __init__(self, data):
        self.data = data
        self.children = []

def maxSumNode(tree):
    if tree==None:
        return None,0
    
    ans=tree
    sum=tree.data
    for child in tree.children:
        sum=sum+child.data
    
    for child in tree.children:
        child_max,child_sum=maxSumNode(child)
        if child_sum>sum:
            sum=child_sum
            ans=child_max
            
    return ans,sum        

def createLevelWiseTree(arr):
    root = treeNode(int(arr[0]))
    q = [root]
    size = len(arr)
    i = 1
    while i<size:
        parent = q.pop(0)
        childCount = int(arr[i])
        i += 1
        for j in range(0,childCount):
            temp = treeNode(int(arr[i+j]))
            parent.children.append(temp)
            q.append(temp)
        i += childCount
    return root

# Main
arr = list(int(x) for x in input().strip().split(' '))
tree = createLevelWiseTree(arr)
temp, tempSum = maxSumNode(tree)
print(temp.data)
```

### NEXT LARGEST

```python
class treeNode:
    def __init__(self, data):
        self.data = data
        self.children = []

def nextLargest(tree, n):
    if tree==None:
        return None,0
    
    ans=tree
    if tree.data>n:
        diff=tree.data-n
    else:
        diff=1000000    
    
    for child in tree.children:
        child_ans,child_diff=nextLargest(child, n)
        if child_diff<diff:
            ans=child_ans
            diff=child_diff
            
    return ans,diff    

def createLevelWiseTree(arr):
    root = treeNode(int(arr[0]))
    q = [root]
    size = len(arr)
    i = 1
    while i<size:
        parent = q.pop(0)
        childCount = int(arr[i])
        i += 1
        for j in range(0,childCount):
            temp = treeNode(int(arr[i+j]))
            parent.children.append(temp)
            q.append(temp)
        i += childCount
    return root

# Main
n = int(input())
arr = list(int(x) for x in input().strip().split(' '))
tree = createLevelWiseTree(arr)
ans,diff=nextLargest(tree, n)
print(ans.data)
```
