## KRUSKAL'S ALGORITHM

```python
class Edge:
    def __init__(self,v1,v2,w):
        self.src=v1
        self.dest=v2
        self.weight=w
        
    def __lt__(self,other):
        return self.weight<other.weight   
    
    def __repr__(self):
        return str(self.src)+' '+str(self.dest)+' '+str(self.weight)

def Kruskals(V,E,edges):
    parrent_array=[i for i in range(V)]
    MST=[]
    count,i=0,0
    while (count<(V-1)):
        ele=edges[i]
        
        p_src=ele.src
        arr_ele=parrent_array[ele.src]
        while arr_ele!=p_src:
            p_src=parrent_array[arr_ele]
            arr_ele=parrent_array[p_src]
                
        p_dest=ele.dest
        arr_ele=parrent_array[ele.dest]
        while arr_ele!=p_dest:
            p_dest=parrent_array[arr_ele]
            arr_ele=parrent_array[p_dest]   

        if p_src!=p_dest:
            MST.append(ele)
            if p_src<p_dest:
                parrent_array[p_dest]=parrent_array[p_src]
            else:
                parrent_array[p_src]=parrent_array[p_dest]                
            count+=1
        i+=1
        print(parrent_array)
        
    return MST        
    
V,E=[int(x) for x in input().split()]
edges=[]
for i in range(E):
    curr=[int(x) for x in input().split()]
    v1,v2,w=curr[0],curr[1],curr[2]
    edge=Edge(v1,v2,w)
    edges.append(edge)    
edges.sort()
li=Kruskals(V,E,edges)
print()
for ele in li:
    if ele.src>ele.dest:
        print(ele.dest,ele.src,ele.weight)
    else:
        print(ele.src,ele.dest,ele.weight)
```

## PRIMS ALGORITHM

## DIJKSTRAS ALGORITHM

```python
import sys
class Graph:
    def __init__(self,nVertices):
        self.vertices=nVertices
        self.adjacentMatrix=[[0 for j in range(nVertices)] for i in range(nVertices)]
        
    def addEdge(self,v1,v2,w):
        self.adjacentMatrix[v1][v2]=w
        self.adjacentMatrix[v2][v1]=w
        return
        
    def removeEdge(self,v1,v2):
        if not(self.containsEdge):
            return 
        self.adjacentMatrix[v1][v2]=0
        self.adjacentMatrix[v2][v1]=0
        return
    
    def __getMinVertex(self,visited,distance):
        min_vertex=-1
        for i in range(self.vertices):
            if (visited[i] is False) and (min_vertex==-1 or distance[i]<distance[min_vertex]):
                min_vertex=i
        return min_vertex                     
    
    def Dijkstras(self):
        visited=[False]*(self.vertices)
        distance=[0]
        distance.extend([sys.maxsize]*(self.vertices-1))
        for i in range(self.vertices-1):
            min_vertex=self.__getMinVertex(visited,distance)
            visited[min_vertex]=True
            for j in range(self.vertices):
                if ((self.adjacentMatrix[min_vertex][j]!=0) and (visited[j] is False)):
                    if distance[j]>(distance[min_vertex]+self.adjacentMatrix[min_vertex][j]):
                        distance[j]=distance[min_vertex]+self.adjacentMatrix[min_vertex][j] 
                        
        print()
        for i in range(self.vertices):
            print(i,distance[i])
        
        return                
    
    def containsEdge(self,v1,v2):
        return self.adjacentMatrix[v1][v2]==1
    
    def __str__(self):
        return str(self.adjacentMatrix)

V,E=[int(x) for x in input().split()]
g=Graph(V)
for i in range(E):
    v1,v2,w=[int(x) for x in input().split()]
    g.addEdge(v1,v2,w)

g.Dijkstras() 
```
